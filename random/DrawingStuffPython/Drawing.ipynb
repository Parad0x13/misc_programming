```python
import numpy as np
import matplotlib.pyplot as plt

# z = z**2 + c
def orbits(c, N = 16):
    z = complex(0, 0)
    for n in range(N):
        z = z**2 + c
        if abs(z) > 2.0: return False
    return True

a = complex(-2, 1)
b = complex(1, -1)
w = 200
h = 320
xDelta = (b.real - a.real) / w
yDelta = (b.imag - a.imag) / h
for y in range(h):
    for x in range(w):
        xPrime = a.real + (xDelta * x)
        yPrime = a.imag + (yDelta * y)
        if orbits(complex(xPrime, yPrime)) == False:
            plt.scatter(xPrime, yPrime, s = 10, c = (0, 0, 0))
plt.show()
```
